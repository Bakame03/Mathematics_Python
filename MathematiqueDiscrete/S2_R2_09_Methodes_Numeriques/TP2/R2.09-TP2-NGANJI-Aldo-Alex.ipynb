```python
def f(a, n):
    # Initialisation de la suite avec x1 = 1
    x = 1.0
    resultats = [x]
    b = a + 1
    
    for i in range(1, n):
        # Calcul du terme suivant : x_{n+1} = b*x_n - a
        x = b * x - a
        resultats.append(x)
        
    return resultats
```

```python
# Test de la fonction avec les paramètres:
a = 1023.1
n = 10

valeurs_suite = f(a, n)

print(f"Question 3 - Termes de la suite pour a = {a} :")
for i, val in enumerate(valeurs_suite):
    print(f"x_{i+1} = {val}")
```

```python
def dif(a, n):
    y0 = 0.5
    ya = y0
    yb = y0
    differences = []
    
    for i in range(n):
        # Formule (a)
        ya = (a + 1) * ya - a * (ya^2)
        # Formule (b)
        yb = (a + 1) * yb - a * (yb) * yb
        # Stockage de la différence
        differences.append(ya - yb)
    
    # Création du graphique
    return list_plot(differences, plotjoined=True, title="Différences entre les deux expressions", 
                     axes_labels=['n', 'ya - yb'])

# Affichage du graphique pour n=100
show(dif(3.0, 100))
```

```python
def Yn(n):
    a = 3.0
    # Initialisation des 5 variantes avec y0 = 0.5
    y_vals = [0.5] * 5
    
    for i in range(n):
        y = y_vals 
        
        y_next = [
            (a + 1) * y[0] - a * (y[0]^2),        # (a)
            (a + 1) * y[1] - a * (y[1]) * y[1],   # (b)
            ((a + 1) - a * y[2]) * y[2],          # (c)
            a * y[3] + (1 - a * y[3]) * y[3],     # (d)
            y[4] + a * (y[4] - y[4]^2)            # (e)
        ]
        y_vals = y_next
        
    return y_vals

print(f"Question 3 - Valeurs de y_100 pour les 5 expressions :")
resultats_100 = Yn(100)
for i, val in enumerate(resultats_100):
    print(val)
```
