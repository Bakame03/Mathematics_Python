```python
x = var('x')
equation = x^2 - x - 1 == 0

# Résolution exacte
solutions = solve(equation, x)
# On récupère la solution positive (la deuxième dans la liste)
sol_positive = solutions[1].right()

# Valeur approchée
valeur_approchee = sol_positive.n()

print("Question 3 - Le Nombre d'Or :")
print(f"Valeur exacte : {sol_positive}")
print(f"Valeur approchée : {valeur_approchee}")

# Conclusion : La suite u_n semble converger vers le nombre d'or
```

```python
def f(a, n):
    # Initialisation de la suite avec x1 = 1
    x = 1.0
    resultats = [x]
    b = a + 1
    
    for i in range(1, n):
        # Calcul du terme suivant : x_{n+1} = b*x_n - a
        x = b * x - a
        resultats.append(x)
        
    return resultats
```

```python
# Test de la fonction avec les paramètres:
a = 1023.1
n = 10

valeurs_suite = f(a, n)

print(f"Question 3 - Termes de la suite pour a = {a} :")
for i, val in enumerate(valeurs_suite):
    print(f"x_{i+1} = {val}")
```
