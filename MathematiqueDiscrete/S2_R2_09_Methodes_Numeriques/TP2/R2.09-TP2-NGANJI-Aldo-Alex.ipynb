```python
# Initialisation avec une valeur numérique (u0 = 1)
u_num = 1.0

print("Question 2 - Calcul numérique des 20 premiers termes :")
for i in range(20):
    u_num = sqrt(u_num + 1)
    print(u_num)

# Conjectures : 
# 1. Sens de variation : La suite semble croissante
# 2. Limite : La suite semble converger vers environ 1.618
```

```python
x = var('x')
equation = x^2 - x - 1 == 0

# Résolution exacte
solutions = solve(equation, x)
# On récupère la solution positive (la deuxième dans la liste)
sol_positive = solutions[1].right()

# Valeur approchée
valeur_approchee = sol_positive.n()

print("Question 3 - Le Nombre d'Or :")
print(f"Valeur exacte : {sol_positive}")
print(f"Valeur approchée : {valeur_approchee}")

# Conclusion : La suite u_n semble converger vers le nombre d'or
```
