```python
def dijkstra(M):
    n = M.nrows()                     
    Dist = [Infinity] * n             
    Dist[0] = 0                       
    marquage = [0] * n                

    while 0 in marquage:              
        mini = Infinity
        
        i = -1
        for k in range(n):
            if marquage[k] == 0 and Dist[k] < mini:
                mini = Dist[k]
                i = k

        if i == -1:                   
            break

        marquage[i] = 1               
        
        for j in range(n):            
            if marquage[j] == 0:
                poids = M[i][j]
                if poids == 0 and i != j:
                    poids = Infinity  
                if Dist[i] + poids < Dist[j]:
                    Dist[j] = Dist[i] + poids
    return Dist
```
